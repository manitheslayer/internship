```python
# importing required libraries
import pandas as pd
import warnings
import time
warnings.filterwarnings("ignore")
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
from selenium.common.exceptions import NoSuchElementException
```

```python
# initating the chrome web driver
driver = webdriver.Chrome("D:\Web Scrapper\chromedriver.exe")

# getting to the url
driver.get("https://github.com/")
time.sleep(5)
```

```python
# finding the element and clicking it
a=driver.find_elements_by_xpath("//summary[@class='HeaderMenu-summary HeaderMenu-link px-0 py-3 border-0 no-wrap d-block d-lg-inline-block']")
a[1].click()
time.sleep(2)
```

```python
# finding the element and clicking it
b=driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
b.click()
time.sleep(5)
```

```python
# finding the url
url = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
```

```python
# creating a list of urls
urls=[d.get_attribute('href') for d in url]
```

```python
# defining funcitons to find elements
def c():
    try:
        ab=driver.find_element_by_xpath("//p[@class='f4 my-3']")
        about.append(ab.text)
        
        
    except NoSuchElementException:
        about.append("-")

        
def e():
    try:
        bc=driver.find_elements_by_xpath("//span[@class='Counter']")
        contribution.append(bc[(len(bc)-1)].text)
        
        
    except NoSuchElementException:
        contribution.append("-")      
        
        
def f():
    try:
        cd=driver.find_element_by_xpath("//strong[@class='mr-2 flex-self-stretch']")
        name.append(cd.text)
        
        
    except NoSuchElementException:
        name.append("-")   
        
        
        
def g():
    try:
        ef=driver.find_elements_by_xpath("//ul[@class='list-style-none']")
        language.append(ef[(len(ef)-1)].text.replace("\n"," "))
        
        
    except (NoSuchElementException,IndexError):
        language.append("-")         
```

```python
# creating a empty list for storing data
about=[]
contribution=[]
name = []
language=[]
# calling the functions using for loop
for i in urls:
    driver.get(i)
    time.sleep(3)
    c()
    e()
    f()
    g()
    
```

```python
# creating a dataframe with collected data
df = pd.DataFrame()
df["Name of Repository"] = name
df["Repository Description"] = about
df["Contributors Count"] = contribution
df["Language used"] = language
df
```
